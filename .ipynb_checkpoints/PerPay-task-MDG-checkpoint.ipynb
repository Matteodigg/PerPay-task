# Predicting Default and Liquidation Rates for Perpay Marketplace Loans

## Introduction

### Business Context

Perpay offers consumers the opportunity to purchase items through installment loans on its marketplace platform, facilitating both credit building and access. Accurately forecasting the future default and liquidation rates of these loans is critical for assessing the financial health of the business, managing risk, and making informed strategic decisions.

### Data Description

This analysis poses interesting challenges due to the dataset's structure. The dataset includes the payment status of a subset of loans from the Perpay Marketplace as of May 9th, 2023, and consists of 620 records, each representing a cohort of loans that entered repayment on a specific date. For each `Date of Repayment`, we have:

- **Number of Loans**: Total loans that began repayment on that date.
- **Default (as of 2023-05-09)**: Number of loans that had defaulted by May 9th, 2023.
- **Liquidation (as of 2023-05-09)**: Number of loans that had been fully paid off by May 9th, 2023.

The dataset has several key characteristics essential to informing our strategy:

- **Loan Maturity**: Loans ultimately reach a terminal state of either default or liquidation. A record is considered "Mature" when the sum of Default and Liquidation equals the "Number of Loans" for that record. If not, this indicates "in-flight" or "active" loans (loans that are still in repayment).
- **Default**: A loan defaults after 120 days of delinquency. This period starts from the last installment made by the customer. For example, if a customer fails to make any installment payments, the loan is classified as defaulted 120 days from the Date of Repayment.
- **Loan Duration**: Loans typically last around 90 days but can vary due to overpayments, underpayments, or extensions.

### Objective

The goal is to develop a predictive model that forecasts the default and liquidation rates of loans _n_ days from May 9th, 2023, for any given _n_. The model must account for the unique challenges presented by the dataset, including censored data, potential trends over time, and a lack of granular features. The model must:

- **Accommodate aggregate cohort-level data**: We lack individual loan-level data.
- **Consider competing risks**: Default, liquidation and in-flight are mutually exclusive statuses with Default and Liquidation being terminal statuses.
- **Incorporate business rules**: No defaults can occur before 120 days since repayment, rates should not change once a loan has fully matured.

## Modeling Approach

### Overview

Given the aggregate nature of the data and specific business constraints, traditional individual-level survival analysis techniques suitable for censored data may not be directly applicable. Instead, I opt for an aggregate-level modeling strategy using custom functions to fit and capture the cumulative behavior of default and liquidation rates over time.

### Choice of Functions

- **Default Rate Function:** I use an exponential cumulative function for the default rate, ensuring that the default probability is zero before 120 days since repayment, in line with business rules. The cumulative exponential function reflects better the trend of default rates, as noticed during EDA.

- **Liquidation Rate Function:** I model the liquidation rate using a standard logistic function, which captures the S-shaped cumulative behavior observed in the data.

By utilizing these functions, I can effectively model the progression of default and liquidation rates over time while adhering to established business constraints. There is no need to explicitly model in-flight rates, as these can be derived as residuals from the fitted functions for default and liquidation.

This approach allows for a robust representation of both default and liquidation rates, respecting their mutual exclusivity while aligning with business requirements. Furthermore, the models produced are interpretable and can be readily communicated to stakeholders, making adjustments straightforward as circumstances change.

#### Default Rate Function

I used an **exponential cumulative function** for modeling the default rate:

$$
P_{\text{default}}(t) = L_1 \cdot \left(1 - e^{-k_1 \cdot (t - 120)}\right) \quad \text{for } t \geq 120
$$

Where the parameters are defined, for our use case, as:
- **`t`**: Represents the time since repayment, measured in days.
- **`L`**: Represents the maximum cumulative default rate.
- **`k`**: The growth rate for the default probability, determining how quickly the probability rises after day 120.

The function ensures that the default probability is zero before 120 days and provides a smooth, continuous increase after such threshold, reflecting the cumulative nature of defaults.

#### Liquidation Rate Function

For modeling the liquidation rate I used a **standard logistic function**:

$$
f(t) = \frac{L}{1 + e^{-k \cdot (t - t_0)}}
$$

With the following parameters:
- **`t`**: Time since repayment, measured in days.
- **`L`**: Maximum cumulative liquidation rate.
- **`k`**: Growth rate
- **`t0`**: Inflection point where the cumulative liquidation rate reaches half of its maximum value (L).

## Data Splitting

To evaluate the model's predictive performance and ensure its generalizability, I decided to split the data into training and validation sets. A careful splitting strategy is essential although I might not have enough data for a ribust test set. Therefore, I decided to use a 70/30 split strategy only on in-flight data, which are of course the most informative for my model.

### Monthly Cohort Splitting Approach

I adopted a monthly cohort splitting method, in which the split is performed on censored data stratified by each month available in the dataset. below a quick summary of this strategy:

- **Process**:
  - **Group data by Date of Repayment**: Preserves the temporal structure of the data.
  - **Within each month**:
    - **Identify In-flight loans**: Loans that have not yet reached a terminal status.
    - **Randomly sample In-flight loans**: Assign 30% of the records to the validation set.
    - **Assign remaining loans to training set**: Ensures the training data contains both mature and the 70% of the remaining in-flight loans.

I opted for this solution for two main reasons:
  - **Temporal integrity**: Avoids mixing data across different time periods.
  - **Realistic validation set**: The validation set contains in-flight loans only, representing future unseen data to better evaluate the model.

```python
# Add Days Since Repayment
data['Days Since Repayment'] = (snapshot_date - data['Date of Repayment']).dt.days

# Count the in-flight loans
data['In-flight'] = (data['Number of Loans'] - data['Default (as of 2023-05-09)'] - data['Liquidation (as of 2023-05-09)'])

# Calculate rates for each loan status
data['Default Rate'] = data['Default (as of 2023-05-09)'] / data['Number of Loans']
data['Liquidation Rate'] = data['Liquidation (as of 2023-05-09)'] / data['Number of Loans']
data['In-flight Rate'] = data['In-flight'] / data['Number of Loans']

# Define a function to create monthly splits and get training and testing sets
def create_monthly_splits(data, test_fraction=0.3):
    data['Month'] = data['Date of Repayment'].dt.to_period('M')
    train_data = []
    test_data = []

    for month, group in data.groupby('Month'):
        # Filter for in-flight loans
        in_flight_loans = group[group['In-flight'] > 0]
        other_loans = group[group['In-flight'] == 0]

        # Sample from in-flight loans for the test set
        if len(in_flight_loans) > 0:
            test_size = int(len(in_flight_loans) * test_fraction)
            if test_size > 0:
                test_indices = in_flight_loans.sample(n=test_size, random_state=42).index

                # Append the sampled test data
                test_data.append(group.loc[test_indices])

                # Remaining in-flight loans go into the training dataset
                remaining_in_flight = in_flight_loans.drop(test_indices)
                train_data.append(remaining_in_flight)
            else:
                # If test_size is 0, add all in-flight loans to training
                train_data.append(in_flight_loans)
        else:
            # Add all loans to training dataset if no in-flight loans
            train_data.append(group)

        # Add all other loans to the training dataset
        train_data.append(other_loans)

    # Combine the data from all months
    train_data_combined = pd.concat(train_data).reset_index(drop=True)
    test_data_combined = pd.concat(test_data).reset_index(drop=True)

    # Drop the 'Month' column as it's no longer needed
    train_data_combined.drop(columns=['Month'], inplace=True)
    test_data_combined.drop(columns=['Month'], inplace=True)

    return train_data_combined, test_data_combined

# Create training and testing sets
train_data, val_data = create_monthly_splits(data)

print(f"\nTraining dataset size: {len(train_data)}")
print(f"Validation dataset size (only active loans): {len(val_data)}\n")
```

## Model Building on Train Data

```python
# Update 'Age' and other calculations based on training data
train_data['Age'] = train_data['Days Since Repayment']

# Identify mature cohorts where 'In-flight' is zero in training data
mature_train_data = train_data[train_data['In-flight'] == 0]

# Calculate the average default and liquidation rates from mature cohorts
Max_Default_Rate = mature_train_data['Default Rate'].mean()
Max_Liquidation_Rate = mature_train_data['Liquidation Rate'].mean()

print(f"Estimated Maximum Default Rate from training data: {Max_Default_Rate:.4f}")
print(f"Estimated Maximum Liquidation Rate from training data: {Max_Liquidation_Rate:.4f}")
```

### Binning the Data

As highlighted during the exploratory data analysis (EDA), default and liquidation rates exhibit significant fluctuations, particularly for loans that are beyond 150 days since repayment, which are not necessarily fully matured. To mitigate these fluctuations, which are largely influenced by the number of loans in each record, I evaluated two approaches:

1. **Weighted Fitting**: When fitting the logistic functions, I considered weighting each data point by the number of loans in the cohort. This could be accomplished using inverse weighting in the `scipy curve_fit()` function with `sigma=1/Number of Loans`. By applying the inverse of the number of loans, cohorts with a larger number of loans exert less variance and thus have a greater influence on the fitting process. This alleviates the problem of high variance arising from rates calculated on smaller cohorts.

2. **Binning the Data**: I also explored the option of binning the data every 10 days to reduce volatility and smooth out random fluctuations, particularly in cohorts with smaller loan counts.

Ultimately, I opted for the binning approach, as I believe it serves as a more effective estimator for fitting the function and smoothing out fluctuations in the data.

```python
train_data
```

```python
# Define bin edges from 0 to max age in steps of 10 days
max_age = train_data['Age'].max()
bin_edges = np.arange(0, max_age + 10, 10)

# Assign each data point to a bin
train_data['Age Bin'] = pd.cut(train_data['Age'], bins=bin_edges, right=False)

# Group training data by 'Age Bin' and calculate weighted average rates
binned_train_data = train_data.groupby('Age Bin', observed = False).apply(lambda x: pd.Series({
    'Age Midpoint': x['Age'].mean(),
    'Number of Loans': x['Number of Loans'].sum(),
    'Default Count': (x['Default (as of 2023-05-09)']).sum(),
    'Liquidation Count': (x['Liquidation (as of 2023-05-09)']).sum(),
}), include_groups=False).reset_index()

# Calculate binned rates
binned_train_data['Default Rate'] = binned_train_data['Default Count'] / binned_train_data['Number of Loans']
binned_train_data['Liquidation Rate'] = binned_train_data['Liquidation Count'] / binned_train_data['Number of Loans']
binned_train_data['In-flight Rate'] = 1 - binned_train_data['Default Rate'] - binned_train_data['Liquidation Rate']
```

Now I am going to fit the function

```python
binned_train_data
```

### Fitting the Default Rate Model

I use the exponential cumulative function to model the default rate, ensuring defaults occur only after 120 days.

```python
# Define the alternative function for default rates (exponential cumulative function)
def exponential_cumulative_default(t, L, k):
    return np.where(
        t >= 120,
        L * (1 - np.exp(-k * (t - 120))),
        0
    )
```

```python
# Ages and observed rates from binned training data
ages_binned = binned_train_data['Age Midpoint'].values
default_rates_binned = binned_train_data['Default Rate'].values

# Filter data for Age >= 120 days for default rate fitting
default_fit_mask = ages_binned >= 120
ages_default_fit = ages_binned[default_fit_mask]
default_rates_fit = default_rates_binned[default_fit_mask]

# Initial guess parameters and bounds. I used the mean of the default rate as max value
# with a small growth rate given the low variance in the dataset
initial_guess_default = [Max_Default_Rate, 0.001]
bounds_default = ([0, 0], [1, np.inf])

# Fit the model
popt_default, _ = curve_fit(
    exponential_cumulative_default,
    ages_default_fit,
    default_rates_fit,
    p0=initial_guess_default,
    bounds=bounds_default,
    absolute_sigma=True,
    maxfev=10000
)

L_default, k_default = popt_default

print(f"\nFitted parameters for Exponential Cumulative Default Rate (Training Data):")
print(f"L = {L_default:.4f}, k = {k_default:.6f}")
```

### Fitting the Liquidation Rate model

```python
# Define the logistic function for liquidation rates
def logistic_function(t, L, k, t0):
    return L / (1 + np.exp(-k * (t - t0)))
```

```python
# Observed rates and weights
liquidation_rates_binned = binned_train_data['Liquidation Rate'].values

# Initial parameters and bounds
initial_guess_liquidation = [Max_Liquidation_Rate, 0.001, np.median(ages_binned)]
bounds_liquidation = ([0, 0, ages_binned.min()], [1, np.inf, ages_binned.max()])

# Fit the model
popt_liquidation, _ = curve_fit(
    logistic_function,
    ages_binned,
    liquidation_rates_binned,
    p0=initial_guess_liquidation,
    bounds=bounds_liquidation,
    absolute_sigma=True,
    maxfev=10000
)

L_liquidation, k_liquidation, t0_liquidation = popt_liquidation

print(f"\nFitted parameters for Liquidation Rate (Training Data):")
print(f"L = {L_liquidation:.4f}, k = {k_liquidation:.6f}, t0 = {t0_liquidation:.2f}")
```

### Visualizing the Fitted Model

Now I can visualize the fitted models using the original data (not binned) against the observed data to assess the model's goodness and understand how well it captures the underlying trends. The plot also includes the in-flight rate calculated from the residuals of the two functions.

```python
# Define age vector
age_range = train_data['Age'].values

# Compute fitted rates
fitted_default_rates = exponential_cumulative_default(age_range, L_default, k_default)
fitted_liquidation_rates = logistic_function(age_range, L_liquidation, k_liquidation, t0_liquidation)
fitted_inflight_rates = 1 - fitted_default_rates - fitted_liquidation_rates

# Ensure rates are within [0,1] in case of numeric instability
fitted_default_rates = np.clip(fitted_default_rates, 0, 1)
fitted_liquidation_rates = np.clip(fitted_liquidation_rates, 0, 1)
fitted_inflight_rates = np.clip(fitted_inflight_rates, 0, 1)

# Plotting
plt.figure(figsize=(12, 8))
plt.scatter(ages_binned, default_rates_binned, label='Binned Default Rate', color='red', alpha=0.7)
plt.plot(age_range, fitted_default_rates, label='Fitted Default Rate', color='red')

plt.scatter(ages_binned, liquidation_rates_binned, label='Binned Liquidation Rate', color='green', alpha=0.7)
plt.plot(age_range, fitted_liquidation_rates, label='Fitted Liquidation Rate', color='green')

plt.scatter(ages_binned, binned_train_data['In-flight Rate'].values, label='Binned In-flight Rate', color='blue', alpha=0.7)
plt.plot(age_range, fitted_inflight_rates, label='Fitted In-flight Rate', color='blue')

plt.xlabel('Age (Days since Date of Repayment)')
plt.ylabel('Rate')
plt.title('Binned Observed vs Fitted Rates Over Time (Training Data)')
plt.legend()
plt.grid(True)
plt.show()
```

## Validation on Test Data

I apply the fitted models to the validation data to predict default and liquidation rates.

```python
# Update 'Age' and calculate rates in validation data
val_data['Age'] = val_data['Days Since Repayment']

val_data['Default Rate'] = val_data['Default (as of 2023-05-09)'] / val_data['Number of Loans']
val_data['Liquidation Rate'] = val_data['Liquidation (as of 2023-05-09)'] / val_data['Number of Loans']
val_data['In-flight Rate'] = val_data['In-flight'] / val_data['Number of Loans']

# Predict default and liquidation rates using the model
def predict_rates(val_data):
    ages = val_data['Age'].values

    # Predict cumulative rates
    predicted_default_rates = exponential_cumulative_default(ages, L_default, k_default)
    predicted_liquidation_rates = logistic_function(ages, L_liquidation, k_liquidation, t0_liquidation)

    # Again rates between [0,1] for numeric instability
    predicted_default_rates = np.clip(predicted_default_rates, 0, 1)
    predicted_liquidation_rates = np.clip(predicted_liquidation_rates, 0, 1)

    # Normalize rates only if total exceeds 1
    total_rates = predicted_default_rates + predicted_liquidation_rates
    over_one_mask = total_rates > 1
    predicted_default_rates[over_one_mask] /= total_rates[over_one_mask]
    predicted_liquidation_rates[over_one_mask] /= total_rates[over_one_mask]

    return predicted_default_rates, predicted_liquidation_rates

# Get predictions on validation data
val_data['Predicted Default Rate'], val_data['Predicted Liquidation Rate'] = predict_rates(val_data)
```

### Evaluating Model Performance

I calculate the Mean Squared Error (MSE) between observed and predicted rates to quantify prediction errors. These MSE values indicate a very good fit of the model:

```python
# Compute MSE for default rates
mse_default = mean_squared_error(val_data['Default Rate'], val_data['Predicted Default Rate'])
print(f"\nMean Squared Error for Default Rate: {mse_default:.6f}")

# Compute MSE for liquidation rates
mse_liquidation = mean_squared_error(val_data['Liquidation Rate'], val_data['Predicted Liquidation Rate'])
print(f"Mean Squared Error for Liquidation Rate: {mse_liquidation:.6f}")
```

To further evaluate the model's performance, I create a calibration plot comparing the observed vs. predicted probabilities for each state in the validation set. Rates are reported on different plots due to difference on the scale:

```python
# Calibration Plot for Default Rate
plt.figure(figsize=(8, 6))
plt.scatter(val_data['Default Rate'], val_data['Predicted Default Rate'], alpha=0.7)
plt.plot([0, val_data['Default Rate'].max()], [0, val_data['Default Rate'].max()], 'r--')
plt.xlabel('Observed Default Rate')
plt.ylabel('Predicted Default Rate')
plt.title('Calibration Plot for Default Rate')
plt.grid(True)
plt.show()

# Calibration Plot for Liquidation Rate
plt.figure(figsize=(8, 6))
plt.scatter(val_data['Liquidation Rate'], val_data['Predicted Liquidation Rate'], alpha=0.7)
plt.plot([0, val_data['Liquidation Rate'].max()], [0, val_data['Liquidation Rate'].max()], 'r--')
plt.xlabel('Observed Liquidation Rate')
plt.ylabel('Predicted Liquidation Rate')
plt.title('Calibration Plot for Liquidation Rate')
plt.grid(True)
plt.show()
```

The model offers a solid approximation of active loans and effectively captures the observed probabilities in the censored data. However, the calibration plot indicates that the model struggles to account for fluctuations in default rates. Improved tuning of the model parameters may help alleviate these issues and enhance predictive accuracy.

## Predicting Rates

In this section, I will predict the rates for the data at the three requested time points. I have provided two versions of the final dataset: one containing the raw predictions and another that incorporates logic to address the violations encountered when running the dataset through the `submission-validator`.

First, I will present the code for generating the raw predictions, followed by an explanation of the model's limitations in relation to the real data and how we can resolve these issues:

```python
def predict_raw_counts_at_dates(future_dates):
    predictions = data[['Date of Repayment', 'Number of Loans', 'Default (as of 2023-05-09)', 'Liquidation (as of 2023-05-09)']].copy()
    
    for future_date_str in future_dates:
        future_date = pd.to_datetime(future_date_str)
        n_days_from_snapshot = (future_date - snapshot_date).days
        
        # Calculate future ages
        future_ages = data['Age'] + n_days_from_snapshot
        
        # Predicted cumulative rates at future ages
        predicted_default_rates = exponential_cumulative_default(future_ages, L_default, k_default)
        predicted_liquidation_rates = logistic_function(future_ages, L_liquidation, k_liquidation, t0_liquidation)
        
        # Ensure rates are within [0,1]
        predicted_default_rates = np.clip(predicted_default_rates, 0, 1)
        predicted_liquidation_rates = np.clip(predicted_liquidation_rates, 0, 1)
        
        # Predicted cumulative counts
        predicted_default_counts = np.round(predicted_default_rates * data['Number of Loans']).astype(int)
        predicted_liquidation_counts = np.round(predicted_liquidation_rates * data['Number of Loans']).astype(int)
        
        # Add columns to predictions DataFrame
        predictions[f'Default (as of {future_date_str})'] = predicted_default_counts
        predictions[f'Liquidation (as of {future_date_str})'] = predicted_liquidation_counts
    
    # Predict fully mature counts using maximum cumulative rates
    predicted_default_counts_full = np.round(Max_Default_Rate * data['Number of Loans']).astype(int)
    predicted_liquidation_counts_full = np.round(Max_Liquidation_Rate * data['Number of Loans']).astype(int)
    
    # Add fully mature counts to predictions DataFrame
    predictions['Default (fully mature)'] = predicted_default_counts_full
    predictions['Liquidation (fully mature)'] = predicted_liquidation_counts_full
    
    return predictions

# future dates for prediction
future_dates = ['2023-06-08', '2023-07-08', '2023-08-07']

# define age on the original dataset
data['Age'] = (snapshot_date - data['Date of Repayment']).dt.days

# get predictions
predictions_raw = predict_raw_counts_at_dates(future_dates)

# select columns to match desired output
columns_order = ['Date of Repayment', 'Number of Loans', 'Default (as of 2023-05-09)', 'Liquidation (as of 2023-05-09)']
for future_date_str in future_dates:
    columns_order.append(f'Default (as of {future_date_str})')
    columns_order.append(f'Liquidation (as of {future_date_str})')
columns_order.append('Default (fully mature)')
columns_order.append('Liquidation (fully mature)')

predictions_raw = predictions_raw[columns_order]

# Save to CSV
predictions_raw.to_csv('predictions_raw.csv', index=False)
```

Upon examining a snapshot of the raw predictions, it becomes evident that applying the model across all provided data presents several problems that violate business constraints:

```python
filtered_predictions = predictions_raw[predictions_raw['Date of Repayment'].isin(pd.to_datetime(['2021-08-31' ,'2022-10-27', '2022-11-25', '2022-12-05']))]
display(filtered_predictions)
```

- **Decreasing Counts of Defaults and Liquidations Over Time**: Defaults and liquidations are cumulative metrics and should not decrease over time. However, the model estimates probabilities based on a cumulative function in relation to the days since the snapshot, without considering the distribution of the initial statuses. While both the default and liquidation functions perform well for data close to the snapshot date (when the counts of defaults and liquidations are close to zero) they struggle with older active loans (see dataset above).

- **Calculation of Fully Mature Loans**: I should provide final predictions regarding the rates for fully matured loans. In the previous function, I used the `Max_Default_Rate` and `Max_Liquidation_Rate`, which represent the asymptotic values of the two functions indicated by the  `L` parameter provided as initial guess in the logistic and exponential models. This is a valid approach since our functions will converge to these values eventually. However, this method may not always be effective.

In conclusion, this dataset will likely fail all the criteria outlined in the submission validator code, though these failures are expected and justified.

> **NOTE**
> **The raw predictions demonstrate strong performance on loans where the days since repayment are less than 120 days.**
> For fully matured loans, I replicated their values from the snapshot date across the other columns for the new requested time points.

To address these issues and align the dataset with business constraints, I must resolve the following challenges:

1. **Ensure counts do not decrease over time**: Both default and liquidation counts should be non-decreasing over time for each cohort, in accordance with the previous snapshot.
2. **Maintain observed ratios**: The proportions of defaults and liquidations should evolve logically, reflecting the trends predicted by the model without violating cumulative properties.
3. **Prevent counts from exceeding total loans**: The sum of defaults and liquidations should not exceed the total number of loans in any cohort.

To mitigate these limitations, I propose the following strategy:

1. **Utilize remaining in-flight loans for adjustment**:
   - **Calculate remaining loans**: At each future date, compute the number of loans remaining (in-flight) after accounting for cumulative defaults and liquidations up to that point.
   - **Distribute remaining loans proportionally**: Adjustments required to ensure total counts do not exceed the total number of loans should be made using the remaining in-flight loans, distributing them proportionally according to the model's predicted rates.

2. **Enforce non-decreasing counts**:
   - **Cumulative maximum function**: Ensure that the predicted counts at each future date are at least as high as both the observed counts at the snapshot date and the counts predicted at earlier dates.
   - **Separate adjustments**: When making adjustments to prevent exceeding total loans, avoid reducing any counts below their previous values.

3. **Proportional scaling without reducing counts**:
   - **Adjust only the incremental part**: If adjustments are needed, scale only the incremental predicted counts (the increase from the previous date), ensuring that base counts remain unaffected.

4. **Immediate update for fully mature cohorts**:
   - **Check for full maturity**: At each future date, verify if the sum of predicted defaults and liquidations equals the total number of loans. If the cohort reaches full maturity at any point, set the fully mature counts to the current predicted counts.

All these solutions are implemented in the following code, which, while not the most elegant, is effective:

```python
def predict_counts_at_dates(future_dates):
    predictions = data[['Date of Repayment', 'Number of Loans', 'Default (as of 2023-05-09)', 'Liquidation (as of 2023-05-09)']].copy()
    
    # Include 'In-flight' column to identify mature cohorts
    predictions['In-flight'] = data['In-flight']
    
    # Initialize previous predicted counts with observed counts as of snapshot date
    prev_default_counts = predictions['Default (as of 2023-05-09)'].copy()
    prev_liquidation_counts = predictions['Liquidation (as of 2023-05-09)'].copy()
    
    # Initialize fully mature counts
    predictions['Default (fully mature)'] = np.nan
    predictions['Liquidation (fully mature)'] = np.nan
    
    for future_date_str in future_dates:
        future_date = pd.to_datetime(future_date_str)
        n_days_from_snapshot = (future_date - snapshot_date).days
        
        # Calculate future ages
        future_ages = data['Age'] + n_days_from_snapshot
        
        # Predicted cumulative rates at future ages
        predicted_default_rates = exponential_cumulative_default(future_ages, L_default, k_default)
        predicted_liquidation_rates = logistic_function(future_ages, L_liquidation, k_liquidation, t0_liquidation)
        
        # Ensure rates are within [0,1]
        predicted_default_rates = np.clip(predicted_default_rates, 0, 1)
        predicted_liquidation_rates = np.clip(predicted_liquidation_rates, 0, 1)
        
        # Predicted cumulative counts
        predicted_default_counts = np.round(predicted_default_rates * data['Number of Loans']).astype(int)
        predicted_liquidation_counts = np.round(predicted_liquidation_rates * data['Number of Loans']).astype(int)
        
        # Ensure counts do not decrease compared to previous counts
        predicted_default_counts = np.maximum(predicted_default_counts, prev_default_counts)
        predicted_liquidation_counts = np.maximum(predicted_liquidation_counts, prev_liquidation_counts)
        
        # Adjust counts if total exceeds Number of Loans
        total_predicted_counts = predicted_default_counts + predicted_liquidation_counts
        over_total_mask = total_predicted_counts > data['Number of Loans']
        
        if over_total_mask.any():
            # Calculate the excess counts
            excess_counts = total_predicted_counts - data['Number of Loans']
            
            # Distribute excess counts proportionally based on the increments
            default_increments = predicted_default_counts - prev_default_counts
            liquidation_increments = predicted_liquidation_counts - prev_liquidation_counts
            total_increments = default_increments + liquidation_increments
            
            # Avoid division by zero
            total_increments = total_increments.replace(0, np.nan)
            
            # Calculate adjustment factors
            default_adjustment = (default_increments / total_increments) * excess_counts
            liquidation_adjustment = (liquidation_increments / total_increments) * excess_counts
            
            # Adjust counts (ensure we don't decrease counts below previous counts)
            predicted_default_counts[over_total_mask] -= np.floor(default_adjustment[over_total_mask]).astype(int)
            predicted_liquidation_counts[over_total_mask] -= np.floor(liquidation_adjustment[over_total_mask]).astype(int)
            
            # Replace NaN increments with zeros
            predicted_default_counts.fillna(prev_default_counts, inplace=True)
            predicted_liquidation_counts.fillna(prev_liquidation_counts, inplace=True)
        
        # Ensure counts do not decrease compared to previous counts after adjustment
        predicted_default_counts = np.maximum(predicted_default_counts, prev_default_counts)
        predicted_liquidation_counts = np.maximum(predicted_liquidation_counts, prev_liquidation_counts)
        
        # For fully mature cohorts, use the observed counts
        fully_mature_mask = predictions['In-flight'] == 0
        predicted_default_counts.loc[fully_mature_mask] = predictions.loc[fully_mature_mask, 'Default (as of 2023-05-09)']
        predicted_liquidation_counts.loc[fully_mature_mask] = predictions.loc[fully_mature_mask, 'Liquidation (as of 2023-05-09)']
        
        # Update previous counts for next iteration
        prev_default_counts = predicted_default_counts
        prev_liquidation_counts = predicted_liquidation_counts
        
        # Add columns to predictions DataFrame
        predictions[f'Default (as of {future_date_str})'] = predicted_default_counts
        predictions[f'Liquidation (as of {future_date_str})'] = predicted_liquidation_counts
        
        # Check for full maturity
        fully_mature_now = (predicted_default_counts + predicted_liquidation_counts) >= data['Number of Loans']
        # Update fully mature counts if not already set
        not_set_mask = predictions['Default (fully mature)'].isna()
        fully_mature_update_mask = fully_mature_now & not_set_mask
        predictions.loc[fully_mature_update_mask, 'Default (fully mature)'] = predicted_default_counts[fully_mature_update_mask]
        predictions.loc[fully_mature_update_mask, 'Liquidation (fully mature)'] = predicted_liquidation_counts[fully_mature_update_mask]
    
    # For any cohorts not yet fully mature, set fully mature counts using maximum rates
    not_fully_mature_mask = predictions['Default (fully mature)'].isna()
    predicted_default_counts_full = np.round(L_default * data['Number of Loans']).astype(int)
    predicted_liquidation_counts_full = np.round(L_liquidation * data['Number of Loans']).astype(int)
    
    # Ensure counts sum to Number of Loans
    total_fully_mature_counts = predicted_default_counts_full + predicted_liquidation_counts_full
    adjustment_needed = total_fully_mature_counts != data['Number of Loans']
    adjustment = data['Number of Loans'] - total_fully_mature_counts
    predicted_liquidation_counts_full += adjustment  # Adjust liquidation counts
    
    # Ensure counts do not decrease compared to last predictions
    predicted_default_counts_full = np.maximum(predicted_default_counts_full, prev_default_counts)
    predicted_liquidation_counts_full = np.maximum(predicted_liquidation_counts_full, prev_liquidation_counts)
    
    # Update the fully mature counts
    predictions.loc[not_fully_mature_mask, 'Default (fully mature)'] = predicted_default_counts_full[not_fully_mature_mask]
    predictions.loc[not_fully_mature_mask, 'Liquidation (fully mature)'] = predicted_liquidation_counts_full[not_fully_mature_mask]
    
    # Drop the 'In-flight' column before returning
    predictions.drop(columns=['In-flight'], inplace=True)
    
    return predictions

# get predictions
adj_predictions = predict_counts_at_dates(future_dates)

# select columns to match desired output
columns_order = ['Date of Repayment', 'Number of Loans', 'Default (as of 2023-05-09)', 'Liquidation (as of 2023-05-09)']
for future_date_str in future_dates:
    columns_order.append(f'Default (as of {future_date_str})')
    columns_order.append(f'Liquidation (as of {future_date_str})')
columns_order.append('Default (fully mature)')
columns_order.append('Liquidation (fully mature)')

adj_predictions = adj_predictions[columns_order]

# Save to CSV
adj_predictions.to_csv('predictions.csv', index=False)
```
